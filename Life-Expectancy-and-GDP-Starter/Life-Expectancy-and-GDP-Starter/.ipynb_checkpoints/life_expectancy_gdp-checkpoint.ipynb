#### This project aims to analyze data on GDP and life expectancy from the World Health Organization and the World Bank to identify the relationship between GDP and life expectancy of six countries.

Below we have some import statements to allow us to use the appropriate libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

Next, we need to import the data. It is contained in a file titled "all_data.csv".

```python
df = pd.read_csv("all_data.csv")
```

Let's take a look at the first and last few rows of the data.

```python
df.head(10)
```

```python
df.tail(10)
```

We can take a quick look for any null values and the data types with .info()

```python
print(df.info())
```

I want to be able to use the columns without needing the brackets and quotes. I will simply convert the column names below.

```python
df.columns = map(str.lower, df.columns)
def space_replacer(column_name):
    return column_name.replace(" ", "_").replace("(", "").replace(")", "")
df.columns = map(space_replacer, df.columns)
```

I want to learn a little more about this data. Let's take a look at the unique values in each column.

```python
print("The unique countries included in this data set are: ")
for country in df.country.unique():
    print(country)
```

```python
print("The unique years included in this data set are: ")
for year in df.year.unique():
    print(year)
```

I did notice that the GDP is formatted in scientific notation. This makes it a bit hard to compare values.
Let's try to convert this into something that makes more sense to us. I will turn them into values that represent billions of dollars.
So, for example, the value 1.45 would be equal to $1,450,000,000, or 1.45 billion dollars.

```python
df.gdp = df.gdp.apply(lambda x: x / 1000000000)
```

```python
df.gdp = df.gdp.apply(lambda x: round(x, 2))
```

Everything with the categorical data seems ok and we don't seem to have any null-values in the quantitative data.
We can move on to some EDA.

### Exploratory Data Analysis

Let's take a look at some of the summary statistics for this data.

```python
n = 1
t = 2
d = 6
w = 0.8
x_values1 = [t * element + w * n for element in range(len(df.groupby("country").life_expectancy_at_birth_years.mean()))]
n = 2
t = 2
d = 6
w = 0.8
x_values2 = [t * element + w * n for element in range(len(df.groupby("country").life_expectancy_at_birth_years.median()))]
mean_life_expectancy = df.groupby("country").life_expectancy_at_birth_years.mean().sort_values(ascending=False)
median_life_expectancy = df.groupby("country").life_expectancy_at_birth_years.median().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(12,5))
ax.bar(x_values1, mean_life_expectancy, width=0.4, label='Mean Life Expectancy')
ax.bar(x_values2, median_life_expectancy, width=0.4, label='Median Life Expectancy')
ax.set_xticks([(a + b) / 2 for a, b in zip(x_values1, x_values2)])
ax.set_xticklabels(mean_life_expectancy.index, rotation=30)
ax.set_xlabel('Country')
ax.set_ylabel('Life Expectancy at Birth (Years)')
ax.set_title('Life Expectancy by Country')
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
n = 1
t = 2
d = 6
w = 0.8
x_values1 = [t * element + w * n for element in range(len(df.groupby("country").gdp.mean()))]
n = 2
t = 2
d = 6
w = 0.8
x_values2 = [t * element + w * n for element in range(len(df.groupby("country").gdp.median()))]
mean_gdp = df.groupby("country").gdp.mean().sort_values(ascending=False)
median_gdp = df.groupby("country").gdp.median().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(12,5))
ax.bar(x_values1, mean_gdp, width=0.4, label='Mean GDP')
ax.bar(x_values2, median_gdp, width=0.4, label='Median GDP')
ax.set_xticks([(a + b) / 2 for a, b in zip(x_values1, x_values2)])
ax.set_xticklabels(mean_gdp.index, rotation=30)
ax.set_xlabel('Country')
ax.set_ylabel('GDP')
ax.set_title('GDP (in billions) by Country')
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

#### What does this tell us?
We can clearly see that the top 3 countries in terms of their GDP are the "United States of American", "China", and "Germany".
This is true regardless of whether you compare them according to their mean or their median values.

##### Does this higher value in GDP have any correlation to life expectancy?
We can clearly see that the top 3 countries in terms of their life expectancies are "Germany", "Chile", and the "United States of American".
Without digging too deep into the data we can take note that both "Germany" and the "United States of America" are in the top 3 for both categories.
This gives us a clue that the two variables might be related in some way.

Before we continue though we should note that Zimbabwe has a very low value for it's Life Expectancy.

We should check in on the data for Chile as well to make sure it doesn't have any outliers.
It is slightly suspicious that they have such a high Life Expectancy and such a low GDP.

The easiest way to do this would be to just check on the distribution of data for each country and for each variable.

```python
plt.figure(figsize=(14, 6))
plt.subplot(2, 3, 1)
plt.hist(df[df.country == "Chile"].gdp)
plt.xlabel("GDP")
plt.ylabel("Density")
plt.title("Chile")
plt.subplot(2, 3, 2)
plt.hist(df[df.country == "China"].gdp)
plt.xlabel("GDP")
plt.ylabel("Density")
plt.title("China")
plt.subplot(2, 3, 3)
plt.hist(df[df.country == "Germany"].gdp)
plt.xlabel("GDP")
plt.ylabel("Density")
plt.title("Germany")
plt.subplot(2, 3, 4)
plt.hist(df[df.country == "Mexico"].gdp)
plt.xlabel("GDP")
plt.ylabel("Density")
plt.title("Mexico")
plt.subplot(2, 3, 5)
plt.hist(df[df.country == "United States of America"].gdp)
plt.xlabel("GDP")
plt.ylabel("Density")
plt.title("United States of America")
plt.subplot(2, 3, 6)
plt.hist(df[df.country == "Zimbabwe"].gdp)
plt.xlabel("GDP")
plt.ylabel("Density")
plt.title("Zimbabwe")
plt.subplots_adjust(wspace=0.6, hspace=0.6)
plt.show()
```

The shape of the data sets is not too concerning. We do see that the datasets for Zimbabwe and China do see a little more skewed than the others but it doesn't seem to be be pulling the mean values for those countries too far from the median values. So I think the mean values are still a fair way to compare each country to one another.

Let's take a look at the distribution of the Life Expectancy for each country.

```python
plt.figure(figsize=(14, 6))
plt.subplot(2, 3, 1)
plt.hist(df[df.country == "Chile"].life_expectancy_at_birth_years)
plt.xlabel("Life Expectancy")
plt.ylabel("Density")
plt.title("Chile")
plt.subplot(2, 3, 2)
plt.hist(df[df.country == "China"].life_expectancy_at_birth_years)
plt.xlabel("Life Expectancy")
plt.ylabel("Density")
plt.title("China")
plt.subplot(2, 3, 3)
plt.hist(df[df.country == "Germany"].life_expectancy_at_birth_years)
plt.xlabel("Life Expectancy")
plt.ylabel("Density")
plt.title("Germany")
plt.subplot(2, 3, 4)
plt.hist(df[df.country == "Mexico"].life_expectancy_at_birth_years)
plt.xlabel("Life Expectancy")
plt.ylabel("Density")
plt.title("Mexico")
plt.subplot(2, 3, 5)
plt.hist(df[df.country == "United States of America"].life_expectancy_at_birth_years)
plt.xlabel("Life Expectancy")
plt.ylabel("Density")
plt.title("United States of America")
plt.subplot(2, 3, 6)
plt.hist(df[df.country == "Zimbabwe"].life_expectancy_at_birth_years)
plt.xlabel("Life Expectancy")
plt.ylabel("Density")
plt.title("Zimbabwe")
plt.subplots_adjust(wspace=0.6, hspace=0.6)
plt.show()
```

We can proceed with some more visualizations.

Below we will make a scatter plot of the two variables to try to visualize the relationship.

```python
sns.scatterplot(x="gdp", y="life_expectancy_at_birth_years", data=df)
plt.xlabel("GDP")
plt.ylabel("Life Expectancy at Birth (Years)")
plt.title("Life Expectancy and GDP Comparison")
plt.show()
```

There is some shape to the scatter plot.
We can zoom in a bit and see if what appears to be an almost veritcal line towards the 0 on the x-axis changes it's shape at all.

```python
df2 = df[(df.gdp < (df.gdp.max() * 0.5))]
sns.scatterplot(x="gdp", y="life_expectancy_at_birth_years", data=df2)
plt.xlabel("GDP")
plt.ylabel("Life Expectancy at Birth (Years)")
plt.title("Life Expectancy and GDP Comparison")
plt.show()
```
